```python
from google.cloud import compute_v1
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/Users/genial-caster-473015-f0-55ca72882d9e.json"
project_id = "genial-caster-473015-f0"  # Replace with your project ID
zone = "us-east4-b"  # Replace with your desired zone
instance_name = "my-python-project-test-vm"
machine_type = f"zones/{zone}/machineTypes/e2-medium"  # Or another machine type
source_image = "projects/debian-cloud/global/images/family/debian-11"
instance_config = compute_v1.Instance(
    name=instance_name,
    machine_type=machine_type,
    disks=[
        compute_v1.AttachedDisk(
            initialize_params=compute_v1.AttachedDiskInitializeParams(
                source_image=source_image,
                disk_size_gb=20,
            ),
            auto_delete=True,
            boot=True,
        )
    ],
    network_interfaces=[
        compute_v1.NetworkInterface(
            name="global/networks/default",  # Or your custom network
            access_configs=[compute_v1.AccessConfig(name="External NAT", type="ONE_TO_ONE_NAT")],
        )
    ],
    # Add other configurations like metadata, service accounts, etc., as needed
)
instance_client = compute_v1.InstancesClient()
request = compute_v1.InsertInstanceRequest(
    project=project_id,
    zone=zone,
    instance_resource=instance_config,
)
print(f"Creating instance {instance_name} in {zone}...")
operation = instance_client.insert(request=request)
operation.result()  # Wait for the operation to complete
print(f"Instance {instance_name} created successfully.")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

from sshCommands import CloudVM

# 1. Basic Config
service_account_text = widgets.Text(
    description="Service JSON",
    placeholder="Path to your service account file",
    layout=widgets.Layout(width="500px"),
)

project_id_text = widgets.Text(
    description="Project ID",
    placeholder="your-gcp-project-id",
    layout=widgets.Layout(width="400px"),
)

zone_dropdown = widgets.Dropdown(
    options=["us-east4-b", "us-central1-a", "us-central1-b"],
    value="us-east4-b",
    description="Zone",
    layout=widgets.Layout(width="250px"),
)

machine_type_dropdown = widgets.Dropdown(
    options=["e2-medium", "e2-standard-4", "n1-standard-4"],
    value="e2-medium",
    description="Machine",
    layout=widgets.Layout(width="250px"),
)

gpu_type_dropdown = widgets.Dropdown(
    options=[
        ("No GPU", "none"),
        ("NVIDIA T4", "nvidia-tesla-t4"),
        ("NVIDIA L4", "nvidia-l4"),
    ],
    value="none",
    description="GPU type",
    layout=widgets.Layout(width="300px"),
)

gpu_count_int = widgets.BoundedIntText(
    value=1,
    min=1,
    max=8,
    step=1,
    description="#GPU",
    layout=widgets.Layout(width="150px"),
)

packages_text = widgets.Text(
    description="Packages",
    placeholder="torch torchvision",
    layout=widgets.Layout(width="500px"),
)

code_textarea = widgets.Textarea(
    description="Code",
    placeholder="Write Python code here to run on the VM",
    layout=widgets.Layout(width="800px", height="200px"),
)

run_button = widgets.Button(
    description="Run on GPU VM",
    button_style="primary",
)

output_area = widgets.Output()



config_box = widgets.VBox(
    [
        service_account_text,
        project_id_text,
        widgets.HBox([zone_dropdown, machine_type_dropdown]),
        widgets.HBox([gpu_type_dropdown, gpu_count_int]),
        packages_text,
    ]
)

gui = widgets.VBox(
    [
        config_box,
        code_textarea,
        run_button,
        widgets.Label("Output:"),
        output_area,
    ]
)

display(gui)
```
