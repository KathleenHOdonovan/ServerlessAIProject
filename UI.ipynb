```python
import ipywidgets as widgets
from IPython.display import display

gpu_type_dropdown = widgets.Dropdown(
    options=[
        ("No GPU", ""),               
        ("T4", "nvidia-tesla-t4"),
        ("V100", "nvidia-tesla-v100"),
    ],
    value="",
    description="GPU Type:",
)

gpu_count_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=4,
    step=1,
    description="GPU Count:",
)

profile_dropdown = widgets.Dropdown(
    options=[
        ("Small (4vCPU, 15GB)", "small"),
        ("Medium (8vCPU, 30GB)", "medium"),
        ("Large (16vCPU, 60GB)", "large"),
    ],
    value="small",
    description="Profile:",
)

run_button = widgets.Button(
    description="Run on VM",
    button_style="success",  # 'success', 'info', 'warning', 'danger' or ''
)

output = widgets.Output()


ui = widgets.VBox([
    profile_dropdown,
    gpu_type_dropdown,
    gpu_count_slider,
    run_button,
    output,
])

display(ui)
```

```python
from IPython import get_ipython

def on_run_button_clicked(b):
    with output:
        output.clear_output()
        
        profile = profile_dropdown.value
        gpu_type = gpu_type_dropdown.value
        gpu_count = gpu_count_slider.value
        
        print("test output")
        print(f"Profile: {profile}")
        print(f"GPU Type: {gpu_type if gpu_type else 'No GPU'}")
        print(f"GPU Count: {gpu_count}")
        
        code = "print('hello')"
        print(code)

run_button.on_click(on_run_button_clicked)
```
